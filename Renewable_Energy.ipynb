```
# Analytics imports
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
```

```
data  = pd.read_csv("data/annual-change-hydro.csv")
data
```

```
country_data = pd.read_csv("http://geohack.net/gis/wikipedia-iso-country-codes.csv")
country_data
```

```
for index,row in data.iterrows():
  country = country_data.loc[country_data['English short name lower case']==row['Entity']]
  print(country)
  if len(country)>0:
    print(country)
    country.reset_index(inplace=True)
    data.at[index,'ISO_Code'] = country['Alpha-3 code'][0]
```

```
data
```

```
def color_maker(value):
  if value>25:
    return "#4575b4"
  elif 10<=value<=25:
    return "#74add1"
  elif 5<=value<=10:
    return "#abd9e9"
  elif 0<=value<=5:
    return "#e0f3f8"
  elif -5<=value<=0:
    return "#fee090"
  elif -10<=value<=-5:
    return "#fdae61"
  elif -25<=value<=-10:
    return "#f46d43"
  else:
    return "#d73027"
```

```
data['color'] = data['Hydro (TWh growth – sub method)'].apply(color_maker)
```

```
data
```

```
data['ISO_Code'].unique()
```

```
px.choropleth(data, 
              locations = 'ISO_Code',
              color='Hydro (TWh growth – sub method)', 
              animation_frame="Year",
              color_continuous_scale="bluyl",
              locationmode='ISO-3',
              title='Annual change in hydropower generation',
              height=700,
              range_color=(data['Hydro (TWh growth – sub method)'].min(),data['Hydro (TWh growth – sub method)'].max())
             )
```
